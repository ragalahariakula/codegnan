```python
#We will pass with all features (13) and give a native approach and check how metrics are 
#varying

#Define all the feature values as belos
k_values = [1,14] #as we have 13 features

#define two empty lists for metrics
mse = []
r2_scores = []
```

```python
import numpy as np
```

```python
data.feature_names
```

```python
#let's assume k = 2 and check 
k = 2
correlation_matrix = np.abs(np.corrcoef(x.T, y))
#print(correlation_matrix)
top_k_indices = np.argsort(correlation_matrix[-1, :-1])[-k:]
print(top_k_indices)
X_selected = x[:, top_k_indices] #checking the data
#print(X_selected)
#let's see what are those two features
top_features = data.feature_names[top_k_indices]
print(top_features)
```

```python
mse = []
r2_scores = []
k_values = range(1,14)
from sklearn.preprocessing import PolynomialFeatures
for k in k_values:
    # Select the top k features with highest correlation
    correlation_matrix = np.abs(np.corrcoef(x.T, y))
    top_k_indices = np.argsort(correlation_matrix[-1, :-1])[-k:]
    X_selected = x[:, top_k_indices]
    #top_features = data.feature_names[top_k_indices]
    #let' split the data into training and testing
    x_train,x_test,y_train,y_test = train_test_split(X_selected,y,
                                                    test_size=0.3,
                                                    random_state=42)
    
    # Define the degree of the polynomial regression
    degree = 2
    # Create polynomial features
    poly_features = PolynomialFeatures(degree=degree)
    X_train_poly = poly_features.fit_transform(x_train)
    X_test_poly = poly_features.transform(x_test)
    
    #building the model
    poly_model = LinearRegression()
    poly_model.fit(X_train_poly,y_train)
    
    # Make predictions on the testing set
    y_pred = poly_model.predict(X_test_poly)

    # Calculate performance metrics
    ms = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # Store the performance metrics
    mse.append(ms) #mse is empty list
    r2_scores.append(r2) #r2_scores is also empty list
#print(mse)
#print(r2_scores)
    #print(len(mse))
    #print(len(r2_scores))
plt.figure(figsize=(10, 6))
plt.plot(k_values, mse, marker='o', 
         label='MSE')
plt.plot(k_values, r2_scores, marker='o', 
         label='R-squared')
plt.title('Performance Metrics vs. Number of Features (k)')
plt.xlabel('Number of Features (k)')
plt.ylabel('Performance Metric')
plt.legend()
plt.grid(True)
plt.show()
```
