```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("mushrooms.csv")
```

```python
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
for col in df.columns:
    if df[col].dtype=='object':
        df[col]=encoder.fit_transform(df[col])
```

```python
df.info()
```

```python
df
```

```python
x=df.values[:,1:]
y=df.values[:,0]
```

```python
x
```

```python
y
```

```python
y=y.astype(int)
y.dtype
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier().fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
entropy_classifier=DecisionTreeClassifier(criterion='entropy',max_depth=5).fit(x_train,y_train)
```

```python
entropy_classifier.score(x_train,y_train)
```

```python
entropy_classifier.score(x_test,y_test)
```

```python
gini_classifier=DecisionTreeClassifier(criterion='gini',max_depth=5).fit(x_train,y_train)
```

```python
gini_classifier.score(x_train,y_train)
```

```python
gini_classifier.score(x_test,y_test)
```

```python
from sklearn import tree
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tree.plot_tree(gini_classifier,class_names=['p','e'],filled=True)
plt.show()
```

```python
tree.plot_tree(model,class_names=['p','e'],filled=True)
plt.show()
```

```python
tree.plot_tree(entropy_classifier,class_names=['p','e'],filled=True)
plt.show()
```
