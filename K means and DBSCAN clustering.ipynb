```python
from sklearn.datasets import make_blobs
```

```python
# generate sample data
x, y_true = make_blobs(n_samples=500, centers=4,
                       cluster_std=0.6, random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans()
```

```python
kmeans.fit(x)
```

```python
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
import matplotlib.pyplot as plt
```

```python
# original
plt.scatter(x[:, 0], x[:, 1], c=y_true, cmap='viridis',
            alpha=0.5, edgecolors='k')

# cluster centers
plt.scatter(centers[:, 0], centers[:, 1], c='red',
            marker='x', s=200)

# predicted labels
plt.scatter(x[:, 0], x[:, 1], c=labels, cmap='viridis',
            alpha=0.5, edgecolors='k')

plt.title('K-Means (Default Parameters)')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
kmeans_new = KMeans(n_clusters=4,
                    init='k-means++',
                    n_init=10,
                    tol=1e-4,
                    algorithm='auto',
                    random_state=0)
```

```python
kmeans_new.fit(x)
```

```python
labels = kmeans_new.labels_
centers = kmeans_new.cluster_centers_
```

```python
# original
plt.scatter(x[:, 0], x[:, 1], c=y_true, cmap='viridis',
            alpha=0.5, edgecolors='k')

# cluster centers
plt.scatter(centers[:, 0], centers[:, 1], c='red',
            marker='x', s=200)

# predicted labels
plt.scatter(x[:, 0], x[:, 1], c=labels, cmap='viridis',
            alpha=0.5, edgecolors='k')

plt.title('K-Means')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
from sklearn.datasets import make_moons
```

```python
x, y_true = make_moons(n_samples=200, noise=0.1, random_state=42)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.3,
                min_samples=5,
                metric='euclidean',
                algorithm='auto')
```

```python
dbscan.fit(x)
```

```python
labels = dbscan.labels_
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
```

```python
labels = dbscan.labels_
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
```

```python
# original
plt.scatter(x[:, 0], x[:, 1], c=y_true, cmap='viridis',
            alpha=0.5, edgecolors='k')

#predicted labels
plt.scatter(x[:, 0], x[:, 1], c=labels, cmap='viridis',
            alpha=0.5, edgecolors='k')

plt.title (f'DBSCAN (Estimated {n_clusters} clusters)')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
