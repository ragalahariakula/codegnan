Generating non linear data

```python
m=100
x=6*np.random.rand(m,1)
x.shape
```

```python
y=0.5*-x**2+x+2+np.random.rand(m,1)
y.shape
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_model=LinearRegression()
lin_model.fit(x,y)
```

```python
lin_pred=lin_model.predict(x)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print(mean_squared_error(y,lin_pred))
print(r2_score(y,lin_pred))
```

visualization

```python
plt.scatter(x,y,color='blue',label="Input data")
plt.plot(x,lin_model.predict(x),color='red',label="Regression line")
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,include_bias=False)
```

```python
x_poly=poly.fit_transform(x)
```

```python
x[0]
```

```python
x_poly[0]
```

```python
poly_model=LinearRegression()
poly_model.fit(x_poly,y)
```

```python
poly_pred=poly_model.predict(x_poly)
```

```python
print(mean_squared_error(y,poly_pred))
print(r2_score(y,poly_pred))
```

```python
plt.scatter(x,y,color='blue',label="Input data")
plt.plot(x,poly_pred,'.',color='red',label="Regression line")
plt.legend()
plt.show()
```
